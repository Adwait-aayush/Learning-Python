```python
#pandas is used for data processing
#pandas dataframe is a 2d tabular data structure with labelled axes

import pandas as pd;


```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()

# we are importing a dataset named fetch_california_housing from sklearn and then in the next step weassign the dataset to a variable called housing
```

```python
print(type(housing))
```

```python
print(housing)
```

```python
df=pd.DataFrame(housing.data,columns =housing.feature_names)
#the data provided above is very unorganized so what we have to do is convert it into a dataframe and we only want the data is the dataframe so we use housing.data and we want the labels or coloums so we use coloums=housing.feature this will assign the feature values of the housing data as labels of dataframe
```
